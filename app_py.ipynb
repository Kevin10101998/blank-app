```python
import streamlit as st
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

st.title('Análisis del Ciclo de Salmuera')

# Opciones del ciclo de salmuera
opciones_proceso = ["SATURACION", "SEDIMENTACION", "FILTRACION", "INTERCAMBIO", "SALMUERA SALIDA", "INDICE DE ELECTROLISIS DEL SALMUERA"]
proceso = st.selectbox('¿Qué parte del proceso de cloro soda desea Analizar?', opciones_proceso)

# Procesos individuales
if proceso == "SATURACION":
    st.subheader('Opciones de Saturador')
    opcion_saturador = st.selectbox('Seleccione el saturador', ['1', '2'])
    uploaded_file = st.file_uploader("Subir archivo SATURADORES.xlsx", type="xlsx")

    if uploaded_file is not None:
        xls = pd.ExcelFile(uploaded_file)
        nombre_hoja = "SATURADOR" + opcion_saturador
        leer = pd.read_excel(xls, nombre_hoja)

        fechas = leer.iloc[:, 0]
        pHs = leer.iloc[:, 2]
        concentraciones = leer.iloc[:, 3]
        durezas = leer.iloc[:, 5]

        st.line_chart(pd.DataFrame({"Fecha": fechas, "pH": pHs}).set_index("Fecha"))
        st.line_chart(pd.DataFrame({"Fecha": fechas, "Concentraciones": concentraciones}).set_index("Fecha"))
        st.line_chart(pd.DataFrame({"Fecha": fechas, "Dureza CaO": durezas}).set_index("Fecha"))

elif proceso == "SEDIMENTACION":
    st.subheader('Opciones de Sedimentadores')
    opcion_sedimentador = st.selectbox('Seleccione el sedimentador', ['A', 'B', 'C', 'D'])
    uploaded_file = st.file_uploader("Subir archivo SEDIMENTADORES.xlsx", type="xlsx")

    if uploaded_file is not None:
        xls = pd.ExcelFile(uploaded_file)
        nombre_hoja = "SEDIMENTADOR " + opcion_sedimentador
        leer_s = pd.read_excel(xls, nombre_hoja)

        fechas = leer_s.iloc[:, 0]
        concentraciones = leer_s.iloc[:, 4]
        excesos_sodas = leer_s.iloc[:, 6]
        excesos_carbonatos = leer_s.iloc[:, 7]

        st.line_chart(pd.DataFrame({"Fecha": fechas, "Concentraciones": concentraciones}).set_index("Fecha"))
        st.line_chart(pd.DataFrame({"Fecha": fechas, "Exceso de Soda": excesos_sodas}).set_index("Fecha"))
        st.line_chart(pd.DataFrame({"Fecha": fechas, "Exceso de Carbonatos": excesos_carbonatos}).set_index("Fecha"))

elif proceso == "FILTRACION":
    uploaded_file = st.file_uploader("Subir archivo FILTRACION.xlsx", type="xlsx")

    if uploaded_file is not None:
        xls = pd.ExcelFile(uploaded_file)
        leer_f = pd.read_excel(xls, xls.sheet_names[0])

        fechas = leer_f.iloc[:, 0]
        durezas = leer_f.iloc[:, 4]
        excesos_sodas = leer_f.iloc[:, 5]
        excesos_carbonatos = leer_f.iloc[:, 6]

        st.line_chart(pd.DataFrame({"Fecha": fechas, "Dureza CaO": durezas}).set_index("Fecha"))
        st.line_chart(pd.DataFrame({"Fecha": fechas, "Exceso de Soda": excesos_sodas}).set_index("Fecha"))
        st.line_chart(pd.DataFrame({"Fecha": fechas, "Exceso de Carbonatos": excesos_carbonatos}).set_index("Fecha"))

elif proceso == "INTERCAMBIO":
    st.subheader('Opciones de Columnas de Intercambio')
    opcion_intercambio = st.selectbox('Seleccione la columna de intercambio', ['102', '103', '104', '105', 'SE'])
    uploaded_file = st.file_uploader("Subir archivo INTERCAMBIO.xlsx", type="xlsx")

    if uploaded_file is not None:
        xls = pd.ExcelFile(uploaded_file)
        nombre_hoja = "C-" + opcion_intercambio
        leer_i = pd.read_excel(xls, nombre_hoja)

        fechas = leer_i.iloc[:, 0]
        durezas = leer_i.iloc[:, 5]
        concentraciones = leer_i.iloc[:, 3]
        aluminios = leer_i.iloc[:, 6]
        sulfatos = leer_i.iloc[:, 11]

        st.line_chart(pd.DataFrame({"Fecha": fechas, "Dureza CaO": durezas}).set_index("Fecha"))
        st.line_chart(pd.DataFrame({"Fecha": fechas, "Concentraciones": concentraciones}).set_index("Fecha"))
        st.line_chart(pd.DataFrame({"Fecha": fechas, "Aluminio": aluminios}).set_index("Fecha"))
        st.line_chart(pd.DataFrame({"Fecha": fechas, "Sulfato de Sodio": sulfatos}).set_index("Fecha"))

elif proceso == "SALMUERA SALIDA":
    st.subheader('Opciones de Salmuera Desgastada')
    opcion_salmuera = st.selectbox('Seleccione la opción', ['SALMUERA SALIDA', 'DECLORADORES'])
    uploaded_file = st.file_uploader("Subir archivo SALMUERA POBRE.xlsx", type="xlsx")

    if uploaded_file is not None:
        xls = pd.ExcelFile(uploaded_file)

        if opcion_salmuera == 'SALMUERA SALIDA':
            leer_ss = pd.read_excel(xls, "SALMUERA SALIDA")
            fechas = leer_ss.iloc[:, 0]
            pHs = leer_ss.iloc[:, 2]
            concentraciones = leer_ss.iloc[:, 3]

            st.line_chart(pd.DataFrame({"Fecha": fechas, "pH": pHs}).set_index("Fecha"))
            st.line_chart(pd.DataFrame({"Fecha": fechas, "Concentraciones": concentraciones}).set_index("Fecha"))

        elif opcion_salmuera == 'DECLORADORES':
            leer_ss = pd.read_excel(xls, "DECLORADORES")
            fechas = leer_ss.iloc[:, 0]
            pHs = leer_ss.iloc[:, 2]
            cloros = leer_ss.iloc[:, 3]

            df = pd.DataFrame({'pH': pHs, 'Cloro': cloros})
            correlacion = df['pH'].corr(df['Cloro'])
            st.write(f"La correlación entre el pH y el Cloro es: {correlacion.round(5)}")

            st.line_chart(pd.DataFrame({"Fecha": fechas, "pH": pHs}).set_index("Fecha"))
            st.line_chart(pd.DataFrame({"Fecha": fechas, "Cloro": cloros}).set_index("Fecha"))
            st.scatter_chart(pd.DataFrame({"Cloro": cloros, "pH": pHs}))

elif proceso == "INDICE DE ELECTROLISIS DEL SALMUERA":
    uploaded_file_intercambio = st.file_uploader("Subir archivo INTERCAMBIO.xlsx", type="xlsx")
    uploaded_file_salmuera = st.file_uploader("Subir archivo SALMUERA POBRE.xlsx", type="xlsx")

    if uploaded_file_intercambio and uploaded_file_salmuera:
        xls_1 = pd.ExcelFile(uploaded_file_intercambio)
        xls_2 = pd.ExcelFile(uploaded_file_salmuera)

        leer_sulfa = pd.read_excel(xls_1, "C-SE")
        leer_salmu = pd.read_excel(xls_2, "SALMUERA SALIDA")

        fechas = leer_sulfa.iloc[:, 0].tolist()
        sulfatos_entrada = leer_sulfa.iloc[:, 11].tolist()

        sulfatos_salida = []
        for fecha in fechas:
            matching_row = leer_salmu[leer_salmu.iloc[:, 0] == fecha]
            if not matching_row.empty:
                sulfatos_salida.append(matching_row.iloc[0, 6])
            else:
                sulfatos_salida.append(None)

        relacion_sulfatos = []
        for i in range(len(sulfatos_entrada)):
            if sulfatos_salida[i] is not None and not np.isnan(sulfatos_salida[i]) and sulfatos_salida[i] != 0:
                relacion_sulfatos.append(sulfatos_entrada[i] / sulfatos_salida[i])
            else:
                relacion_sulfatos.append(np.nan)

        valid_indices = [i for i, val in enumerate(relacion_sulfatos) if not np.isnan(val)]
        fechas_validas = [fechas[i] for i in valid_indices]
        relacion_sulfatos_validos = [relacion_sulfatos[i] for i in valid_indices]

        if relacion_sulfatos_validos:
            varianza = np.var(relacion_sulfatos_validos)
            desviacion_estandar = np.sqrt(varianza)
            media = np.mean(relacion_sulfatos_validos)
            cota_superior = media + desviacion_estandar
            cota_inferior = media - desviacion_estandar

            st.write(f"Varianza: {varianza:.5f}")
            st.write(f"Desviación estándar: {desviacion_estandar:.5f}")
            st.write(f"Cota superior: {cota_superior:.5f}")
            st.write(f"Cota inferior: {cota_inferior:.5f}")

            df = pd.DataFrame({"Fecha": fechas_validas, "Relación de Sulfatos": relacion_sulfatos_validos})
            st.line_chart(df.set_index("Fecha"))

# Run the Streamlit app
st.write('Hecho por kevin Ormaza')
```
